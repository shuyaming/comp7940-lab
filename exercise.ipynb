```python
# Exercise 3: List Iteration
# Write a program to find all factors of the numbers in the list l
l = [52633, 8137, 1024, 999]

# your code here
for num in l:
    print(f"Factors of {num}:")
    for i in range(1, num + 1):
        if num % i == 0:
            print(f"  {i}")
```
